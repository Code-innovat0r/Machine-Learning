```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Normal distribution
```

```python
# Generating the sample data for the normal distribution normal(1, 0.16)
sample = np.random.lognormal(mean =1, sigma=0.4, size=10000)
shape,loc,scale = scipy.stats.lognorm.fit(sample, floc=0)
```

```python
num_bins = 50
```

```python
counts, edges,patches = plt.hist(sample, num_bins, color='r')
```

```python
# Binomail distribution
```

```python
import seaborn
from scipy.stats import binom
```

```python
data = binom.rvs(n=17,p=0.7, loc=0,size=10000)
```

```python
data
```

```python
ax= seaborn.distplot(data, kde=True, color='pink', hist_kws={"linewidth":22, "alpha":0.77})
```

```python
# Poission Distribution
```

```python
p_data = np.random.poisson(5, 10000)
```

```python
plt.hist(p_data, 16, color='g', rwidth=0.4)
```
