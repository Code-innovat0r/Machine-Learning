```python
import pandas as pd
```

```python
data = pd.read_csv('play_golf.csv')
```

```python
data.head()
```

```python
data
```

```python
data.columns
```

```python
data.columns = ['OUTLOOK', 'TEMPERATURE', 'HUMIDITY', 'WINDY', 'PLAYGOLF']
```

```python
data.head()
```

```python
data['OUTLOOK'].unique()
```

```python
day = ['D1','D2','D3','D4','D5','D6','D7','D8','D9','D10','D11','D12','D13','D14']
```

```python
data['DAY'] = day
```

```python
data
```

```python
out = {'Sunny':0,'Overcast':1,'Rainy':2}
data['OUTLOOK'] = data['OUTLOOK'].map(lambda x:out[x])
```

```python
data.head()
```

```python
data['TEMPERATURE'].unique()
```

```python
temp = {'Hot':0, 'Mild':1, 'Cool':2}
data['TEMPERATURE'] = data['TEMPERATURE'].map(lambda x:temp[x])
```

```python
data.head()
```

```python
data['HUMIDITY'].unique()
```

```python
data['HUMIDITY'] = data['HUMIDITY'].map(lambda x : 0 if x=='High' else 1)
```

```python
data.head()
```

```python
data['WINDY'].unique()
```

```python
data['WINDY'] = data['WINDY'].map(lambda x : 1 if x==True else 0)
data.head()
```

```python
data['PLAYGOLF'] = data['PLAYGOLF'].map(lambda x : 1 if x=='Yes' else 0)
data.head()
```

```python
# importing the decission tree classifier
```

```python
data.iloc[:,4]
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(data.iloc[:,:4], data.iloc[:,4])
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
print(tree.plot_tree(dt))
```

```python
# below code is bit abvance to save the png image of the decision tree
from sklearn.tree import export_graphviz
```

```python
feature_name = ['OUTLOOK', 'TEMPERATURE', 'HUMIDITY', 'WINDY']
out = ['Yes', 'No']
dot = export_graphviz(decision_tree=dt, out_file='golf.dot', feature_names=feature_name, class_names=out)
```
