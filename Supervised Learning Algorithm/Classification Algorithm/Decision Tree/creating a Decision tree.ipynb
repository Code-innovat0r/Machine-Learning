```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score as score
```

```python
score(y_test,y_pred)
```

```python
# In this way you can also increas the graph size
from matplotlib import pyplot as plt
plt.figure(figsize=(80,50))
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
```

```python
# creating the decision tree for the Social_Network_Ads.csv
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
clf1 = make_pipeline(StandardScaler(), DecisionTreeClassifier())
x_train,x_test,y_train,y_test = train_test_split(dataset.iloc[:,2:4], dataset.iloc[:,4], train_size=0.8, random_state=0)
```

```python
clf1.fit(x_train,y_train)
```

```python
y_pred = clf1.predict(x_test)
```

```python
score(y_test, y_pred)
```

```python
plot_tree(clf1[1])
```

```python
from sklearn.datasets import load_diabetes
x,y = load_diabetes(return_X_y=True)
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.9, random_state=0)
```

```python
clf2 = DecisionTreeClassifier(criterion='entropy')
```

```python
clf2.fit(x_train, y_train)
```

```python
y_pred = clf2.predict(x_test)
from sklearn.metrics import confusion_matrix as con
```

```python
score(y_test, y_pred)
```

```python
con(y_test, y_pred)
```
