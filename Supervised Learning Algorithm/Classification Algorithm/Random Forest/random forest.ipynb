```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x=iris.data
```

```python
y=iris.target
```

```python
from sklearn.ensemble import RandomForestClassifier
obj = RandomForestClassifier( random_state=0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_ts,y_tr,y_ts = train_test_split(x,y,random_state=0,train_size=0.75)
```

```python
obj.fit(x_tr,y_tr)
```

```python
y_pr = obj.predict(x_ts)
```

```python
y_pr
```

```python
from sklearn.metrics import accuracy_score as score
```

```python
print('Accuracy score: ', score(y_ts, y_pr))
```

```python
# Doing this prediction using the single decision tree and then compare the accuracy
```

```python
# Entropy
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec = DecisionTreeClassifier(criterion='entropy', splitter='random', random_state=0)
```

```python
y_predict = dec.predict(x_ts)
```

```python
print("Accuracy: ", score(y_ts,y_predict))
```
