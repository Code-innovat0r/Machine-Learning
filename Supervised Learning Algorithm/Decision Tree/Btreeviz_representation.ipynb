```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier , plot_tree
```

```python
data = load_iris()
```

```python
x = data.data
y = data.target
```

```python
clf = DecisionTreeClassifier(max_depth=3)
```

```python
clf.fit(x,y)
```

```python
plot_tree(clf)
```
