```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.shape
```

```python
data
```

```python
# purchased - output
# there are two different classes, one belong to the 0 and other belong to the 1
# so the data have the impurity in it (gini, entropy)
```

```python
x = data.iloc[:,[2,3]].values
```

```python
y = data.iloc[:,4].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train
```

```python
y_train
```

```python
# pre processing the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# importing the decision tree from the sklearn
from sklearn.tree import DecisionTreeClassifier
```

```python
obj = DecisionTreeClassifier(criterion="gini", splitter="random", random_state=0)
```

```python
obj.fit(x_train, y_train)
```

```python
y_pred = obj.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# finding the accuracy and the confusion matric for the predicted data
```

```python
from sklearn.metrics import accuracy_score as score, confusion_matrix as cm
```

```python
print("ACCURACY of the prediction is : ", score(y_test, y_pred))
```

```python
print("Confusion matrix is : ", cm(y_test , y_pred))
```
