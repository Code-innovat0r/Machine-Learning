```python
import pandas as pd
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
# pre-processing of the dataset
```

```python
x = data.iloc[:,[2,3]].values
```

```python
y = data.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=0, test_size=0.25)
```

```python
x_train.shape
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
x_train = scale.fit_transform(x_train)
```

```python
x_test = scale.fit_transform(x_test)
```

```python
x_test
```

```python
# model creation 
# entropy 
from sklearn.tree import DecisionTreeClassifier
obj = DecisionTreeClassifier(criterion='entropy', random_state=0, splitter='random')
```

```python
obj.fit(x_train, y_train)
```

```python
y_pred = obj.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# finding the accuracy and the confusion matric
from sklearn.metrics import accuracy_score as score, confusion_matrix as cs
```

```python
print("Accuracy score is: ", score(y_test, y_pred))
```

```python
print("Confusion matric is: ", cs(y_test, y_pred))
```
