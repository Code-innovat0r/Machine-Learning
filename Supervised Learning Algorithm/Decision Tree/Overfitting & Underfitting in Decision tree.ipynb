```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score as score
```

```python
def analyzer(max_depth):
    data = pd.read_csv('Social_Network_Ads.csv')
    x = data.iloc[:,2:4].values
    y = data.iloc[:,-1].values

    clf = DecisionTreeClassifier(max_depth=max_depth)
    clf.fit(x,y)

    a = np.arange(start=x[:,0].min()-1, stop=x[:,0].max()+1,step=0.1)
    b = np.arange(start=x[:,1].min()-1, stop=x[:,1].max()+1,step=100)
    # if we use the step value of .1 for the b there will be 1350020 data as the start and stop is too far from each other so we take the step of 100

    xx,yy = np.meshgrid(a,b)
    input_array = np.array([xx.ravel(), yy.ravel()]).T

    labels = clf.predict(input_array)

    plt.contourf(xx,yy,labels.reshape(xx.shape),alpha=0.50)
    plt.scatter(x[:,0],x[:,1],c=y)
    
```

```python
# see the overfitting condition majority surrounding is yellow yet the point is classified as purple
analyzer(max_depth = None)
```

```python
# Now the underfitting condition -> no proper distribution of the data
analyzer(max_depth=1)
```
