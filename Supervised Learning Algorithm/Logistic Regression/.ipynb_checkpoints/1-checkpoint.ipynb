```python
# dividing the dataset into input and output
```

```python
# x and y --input and output
# x = data.iloc[:,[2,3]]  # first : is for all rows
x = data.iloc[:,[2,3]].values # convert the data into values
```

```python
x
```

```python
y = data.iloc[:, 4].values # take the prediction columns
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Scaling the value means bringing the value in the limited range
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
# Applying logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state=0)  #you can pass some value or use the defaut value
```

```python
log_reg.fit(x_test, y_test) # model is learning from the dataset
```

```python
y_pred = log_reg.predict(x_test) # y_pred contain those predicted values over the x_test
```

```python
print(y_pred) #printing the pridicted value
```

```python
print(y_test) # printing the original value
```
