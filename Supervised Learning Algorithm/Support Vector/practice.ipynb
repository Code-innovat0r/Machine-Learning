```python
# Support Vector
# Kernal function -> transform input into higher dimensional plane -> seprate the data

```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
# pipeline use for the performin the multiple operation one by one
from sklearn.pipeline import make_pipeline 
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
x = np.array([
    [-1,-1],
    [-2,-3],
    [-2,1],
    [9,-1],
    [3,3],
    [100,100]
])
```

```python
y = np.array([1,1,2,2,3,3])
```

```python
# config is object, make_pipeline will pass the value of config through StandardScaler first then SVC
config = make_pipeline(StandardScaler(), SVC(gamma='auto'))
```

```python
config.fit(x,y)
```

```python
config.predict([[-22,45],[-43,-76],[90,54]])
```

```python
# trying it with the iris dataset
```

```python
from sklearn.datasets import load_iris
```

```python
X,Y = load_iris(return_X_y=True)
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test  =  train_test_split(X,Y, test_size=0.25, random_state=0)
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
```

```python
x_train
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score as score
```

```python
print(score(y_pred, y_test))
```
