```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,2], [2, 2], [2,3], [8,7], [8,8], [80,25]])
```

```python
# visualize the given data on the graph
plt.scatter(x[:,1],x[:,0])
plt.plot
```

```python
# creating for the dbscan object
db = DBSCAN(eps=2 , min_samples=2)
db.fit(x)
```

```python
# viewing the cluster db assigned to the points
db.labels_
```

```python
## Trying out the dbscan on the other datasets to

from sklearn.datasets import make_circles

# create a concentric circle dataset
x, _ = make_circles(n_samples=500, factor=0.5, noise=0.03, random_state=4)

db = DBSCAN(eps=0.1, min_samples=5)
cluster = db.fit_predict(x)

# plotting 
plt.scatter(x[:,0], x[:,1], c=cluster, cmap='viridis', marker='o')
plt.title("The DBSCAN in the concentric circles")
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```

```python
## try to cluster the above data witht the k-mean algorithm

from sklearn.cluster import KMeans

kMean = KMeans(n_clusters=2)

cluster = kMean.fit_predict(x)

plt.scatter(x[:,0],x[:,1], c=cluster, cmap='viridis', marker='o' )
plt.title('KMean clustering result')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```
