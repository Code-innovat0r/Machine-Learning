```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
plt.show()
```

```python
## Lets compare the result with the KMean cluster result

from sklearn.cluster import KMeans

kmean = KMeans(n_clusters=5)
```

```python
cluster = kmean.fit_predict(data)
```

```python
plt.scatter(data[:,0], data[:,1], c=cluster, cmap='viridis', marker='o')
plt.title('Cluster According to K-Mean algorithm')
plt.show()
```

```python
## Lets try the Hierachial cluster over the concentric circles
```

```python
from sklearn.datasets import make_circles

x,_ = make_circles(n_samples=500, random_state=4, factor=0.5, noise=0.03)
model = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster = model.fit_predict(x)
plt.scatter(x[:,0], x[:,1], c= cluster, cmap='viridis', marker='o')
plt.show()
```
