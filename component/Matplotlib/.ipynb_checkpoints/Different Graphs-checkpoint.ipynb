```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.bar([0.25, 0.30, 0.80, 0.90], [20, 30, 10, 40], label = 'population', color ='r', width=0.01)
plt.xlabel('percentage')
plt.ylabel('freq')
plt.show()
```

```python
#  Histogram
```

```python
age = [12, 13, 23, 75, 90, 23, 43]
bins = [0, 10, 20, 30, 100]
```

```python
plt.hist(age, bins=bins, histtype = 'bar', rwidth=0.02)
```

```python
# Scatter
weights = np.array([1,2,3,4,5,6,7,8])
speed = np.array([8,7,6,5,4,3,2,1])
age = np.array([12, 32, 34, 54,67,54,23,32])
```

```python
plt.scatter(weights, speed)
```

```python
# Area plote
```

```python
plt.plot([], [], color = 'g', label = weights, linewidth = 5)
plt.plot([], [], color='r', label = speed, linewidth = 5)
plt.stackplot(weights, speed, age, colors = ['r', 'g'])
```
